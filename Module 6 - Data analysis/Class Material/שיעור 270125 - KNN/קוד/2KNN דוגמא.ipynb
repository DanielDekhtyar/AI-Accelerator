```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

from sklearn.impute import KNNImputer
```

```python
df = sns.load_dataset('penguins')
```

```python
df.isnull().sum()
```

```python
rows_to_check = df[df['sex'].isnull()].index
df[df['sex'].isnull()].head(10)
```

```python
#working just with numbers.
dfn= df.select_dtypes('number')

#scaling 
scaler = StandardScaler()  
scaled_data = scaler.fit_transform(dfn)

#encode
scaled_data = pd.DataFrame(scaled_data,columns= dfn.columns)
sex = df['sex'].astype('category').cat.codes.replace(-1, np.nan)
dfn = pd.concat([scaled_data , sex], 
    axis =1)
dfn.columns.values[-1] = 'sex'
#Knn
imputer = KNNImputer(n_neighbors=3, weights='distance')
emuted  =imputer.fit_transform(dfn)
dfn= pd.DataFrame(emuted ,columns = dfn.columns)
```

```python
dfn.iloc[rows_to_check]
```

```python

mapping = dict(
    enumerate(df['sex'].astype('category').cat.categories)
)

print(mapping)
```

```python
dfn['sex1'] = round(dfn['sex']).map(mapping)
```

```python
df['sex'] = dfn['sex1']
```
