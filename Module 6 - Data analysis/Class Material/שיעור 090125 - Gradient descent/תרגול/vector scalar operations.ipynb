## Section 3: Dot Product Between Two Vectors

### Instructions:
Compute the dot product between the two vectors using the formula:
$ \vec{u} \cdot \vec{v} = \sum_{i=1}^n u_i v_i $

**Example:**
Let $ \vec{u} = [1, 2, 3] $ and $ \vec{v} = [4, 5, 6] $.
$ \vec{u} \cdot \vec{v} = 1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6 = 32 $

**Exercise 3:**
1. $ \vec{u} = [1, 3, 5], \vec{v} = [2, 4, 6] $
   - Compute $ \vec{u} \cdot \vec{v} $.

2. $ \vec{u} = [-1, 0, 3], \vec{v} = [2, -2, 1] $
   - Compute $ \vec{u} \cdot \vec{v} $.
