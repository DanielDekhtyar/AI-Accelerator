## Section 3: Dot Product Between Two Vectors

### Instructions:
Compute the dot product between the two vectors using the formula:
$ \vec{u} \cdot \vec{v} = \sum_{i=1}^n u_i v_i $

**Example:**
Let $ \vec{u} = [2, 3, 4] $ and $ \vec{v} = [1, 0, -1] $.
$ \vec{u} \cdot \vec{v} = 2 \cdot 1 + 3 \cdot 0 + 4 \cdot (-1) = -2 $

**Exercise 3:**
1. $ \vec{u} = [1, 4, 6], \vec{v} = [2, 3, 5] $
   - Compute $ \vec{u} \cdot \vec{v} $.

2. $ \vec{u} = [0, -2, 3], \vec{v} = [1, -1, 2] $
   - Compute $ \vec{u} \cdot \vec{v} $.

## Section 4: Dot Product Between Two Matrices

### Instructions:
Compute the dot product between two matrices using the matrix multiplication formula:
$ A \cdot B = C \text{, where } C_{ij} = \sum_{k=1}^n A_{ik} B_{kj} $

**Example:**
Let $ A = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix} $ and $ B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix} $.
$ A \cdot B = \begin{bmatrix} (1 \cdot 5 + 3 \cdot 7) & (1 \cdot 6 + 3 \cdot 8) \\ (2 \cdot 5 + 4 \cdot 7) & (2 \cdot 6 + 4 \cdot 8) \end{bmatrix} = \begin{bmatrix} 26 & 30 \\ 38 & 44 \end{bmatrix} $

**Exercise 4:**
1. $ A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, B = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix} $
   - Compute $ A \cdot B $.

2. $ A = \begin{bmatrix} 2 & 0 & 1 \\ -1 & 3 & 4 \end{bmatrix}, B = \begin{bmatrix} 3 & 1 \\ 0 & -2 \\ 5 & 4 \end{bmatrix} $
   - Compute $ A \cdot B $.
