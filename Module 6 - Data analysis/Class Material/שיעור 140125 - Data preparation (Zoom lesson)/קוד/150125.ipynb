```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandarScalar, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
#get_dummies
pd.get_dummies(df,columns = ['Sex', 'Embarked'] ).head(1)
pd.get_dummies(df,columns = ['Sex', 'Embarked'] ).info()
#df.drop('Sex_male')
```

```python
#OneHOtEncoder
#1 model
ohe =OneHotEncoder(sparse_output=False) # שינוי  טיפוס התוצאה
```

```python
#fit
encoded = ohe.fit_transform(df[['Sex', 'Embarked']])
print(encoded)
```

```python
#3 result to df
df_e =pd.DataFrame(encoded,columns =ohe.get_feature_names_out(['Sex','Embarked']) )
df_e.head(1)
```

```python
#4 concat both df's
df_new = pd.concat([df,df_e], axis =1)
df_new.head(1)
```

```python
#5 drop original columns
df_new = df_new.drop(['Sex','Embarked'], axis =1)
```
