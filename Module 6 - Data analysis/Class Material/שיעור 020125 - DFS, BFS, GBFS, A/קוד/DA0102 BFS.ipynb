```python
# Breadth-First Search
def bfs(graph, start):
    visited = []
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return visited


# Call the BFS function and print results
visited_nodes_bfs = bfs(graph, "A")


print("BFS - Visited nodes:", visited_nodes_bfs)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Example graph of nodes
graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F", "G"],
    "D": [],
    "E": [],
    "F": [],
    "G": []
}
```

```python
G = nx.DiGraph(graph)

# Positions of the nodes for a tree-like visualization
pos = {"A": (0, 0), "B": (-1, -1), "C": (1, -1), "D": (-2, -2), "E": (0, -2), "F": (0, -3), "G": (2, -2)}

# Graph with node labels
plt.figure(figsize = (4,3))
nx.draw(G, pos, with_labels=True, node_color="#D175A3", font_size=12, font_weight="bold", node_size=1000)
plt.title("Graph")
plt.show()
```

```python
# Node color based on traversal order
def draw_traversal_order(G, pos, traversal_order, title):
    color_map = []
    for node in G:
        if node in traversal_order:
            color_map.append(plt.cm.Greens((traversal_order.index(node) + 1) / len(traversal_order)))
        else:
            color_map.append("lightgray")

    nx.draw(G, pos, with_labels=True, node_color=color_map, font_size=12, font_weight="bold", node_size=1000)
    plt.title(title)
    plt.show()
```

```python
# BFS traversal
visited_nodes_bfs = bfs(graph, "A")
print("BFS - Visited nodes:", visited_nodes_bfs)
draw_traversal_order(G, pos, visited_nodes_bfs, "BFS Traversal")
```
