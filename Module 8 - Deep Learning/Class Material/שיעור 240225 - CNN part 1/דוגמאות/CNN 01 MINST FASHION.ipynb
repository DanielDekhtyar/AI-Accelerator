```python
# Explore CV with the fashion MNIST dataset
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 load
 (train_images, train_labels),(test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# EDA Function Show
num_rows = 2
num_cols = 4
num_images = num_rows * num_cols

classes = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

random_indices = np.random.choice(len(train_images), num_images, replace=False)
#indicies = range(num_images)
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6))

for j in range(num_images):
  idx = random_indices[j] #iindicies[j]
  ax = axes[j // num_cols, j % num_cols]
  ax.imshow(train_images[idx], cmap='gray')
  ax.set_title(classes[train_labels[idx]])
plt.tight_layout()
```

```python
#2 normalize "min max scaler"
train_norm =train_images/255
test_norm =test_images /255
```

```python
train_labels
```

```python
#3 one hot encoding
train_labels =tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
train_labels
```

#Model

```python
#build
model = tf.keras.Sequential([
    #יבואו עוד
    tf.keras.layers.Flatten(),#input_shape=(28,28)
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(10,activation= 'softmax') #(num_of_labels, softmax)

])
```

```python

#compile
model.compile(loss= tf.keras.losses.CategoricalCrossentropy(),
            optimizer = tf.keras.optimizers.Adam(),
              metrics=['accuracy'])


```

```python
#fit
history = model.fit(train_norm,
                    train_labels ,
                    epochs=5,
                    validation_data =(test_norm, test_labels))
```
