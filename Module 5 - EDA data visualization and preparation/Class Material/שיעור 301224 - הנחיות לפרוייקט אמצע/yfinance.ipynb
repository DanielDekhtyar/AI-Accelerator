```python
btc['week_p'] =btc.index.to_period('W')
```

```python
btc.info()
```

```python
btc['month_p'] = btc.index.to_period('M')
```

```python
btc.head(40)
```

```python
btc.groupby('week_p')['Close'].mean()
```

```python
weekly_avg = btc.groupby('week_p')['Close'].mean()
```

```python
monthly_avg = btc.groupby('month_p')['Close'].mean()
```

```python
plt.plot (btc.index, btc['Close']) #:-)
```

```python

plt.plot (monthly_avg.index.to_timestamp(), monthly_avg.values, 
          color = 'blue',
         label ='monthly')
plt.plot (btc.index, btc['Close'],
         color = 'orange',
         label = 'dayly')
plt.plot (weekly_avg.index.to_timestamp(), weekly_avg.values, 
          color = 'green',
         label ='weekly')
plt.legend()
```
