```python
import pandas as pd
```

```python
cereal = pd.read_csv('cereal.csv')
```

```python
cereal.set_index('name',inplace = True)
```

```python
cereal.sort_index(inplace = True)
```

```python
## WHERE
cereal.head(10)
mask =  cereal['mfr']== 'K'
cereal[mask]
```

```python
cereal[mask].shape[0]
```

```python
cereal.where(mask)
```

```python
cereal.where(mask).shape[0]
```

```python
mask2 =  cereal['sugars']== 0
cereal.where(mask & mask2)
```

```python
cereal.where(mask, 'לא משנה')
```

```python
cereal['mfr'].where(mask, 'לא משנה')
```

```python
cereal['mfr'] = cereal['mfr'].where(mask, 'לא משנה')
```
