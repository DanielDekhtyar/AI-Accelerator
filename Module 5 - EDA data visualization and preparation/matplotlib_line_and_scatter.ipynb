 1. Create a function called f3 that takes an argument and multiplies it by 3.
2. Use np.linspace to create an array of 100 numbers between -20 and 20.
3. Plot the array and the results of the function applied to it.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.linspace(-20, 20, 100)
```

```python
def f3(array):
    return array*3
```

```python
plt.plot(array, f3(array))
plt.show()
```


4. Change the line color to green.
5. Change the line style to dashed.
6. Add titles to the axes and the plot.

```python
plt.plot(array, f3(array),c = "green", linestyle = "--")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

6.	Enhance the scatter plot by using petal width as the size of the points and sepal width as the color
7.	In the time left you customize your plot


```python
iris = pd.read_csv("Iris.csv")
```

```python
iris.drop(columns = "Id", inplace = True)
```

```python
sepal_length = iris["SepalLengthCm"]
petal_width = iris["PetalWidthCm"]
```

```python
size = (petal_width**2)*10
color = 1/sepal_length
```

```python
plt.scatter(sepal_length, petal_width, s = size, c = color)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
plt.title("Iris")
plt.show()
```
