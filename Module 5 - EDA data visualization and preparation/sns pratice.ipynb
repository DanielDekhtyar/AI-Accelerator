```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

1. Load the `cereal.csv` file into a pandas DataFrame.  
   Use a heatmap to check for correlations between the columns. Identify:
    - The strongest positive correlation.  
    - The strongest negative correlation.  

```python
cereal = pd.read_csv('datasets/cereal.csv')
```

```python
sns.heatmap(cereal.select_dtypes('number').corr(), cmap="plasma")
plt.show()
```

2. Create a scatter plot in Seaborn to  
visualize the strongest positive correlation you found.

```python
sns.scatterplot(x='potass', y='fiber', data=cereal)
plt.show()
```

3. Update the scatter plot to include:
    - A `hue` based on the column: `mfr`.
    - A `size` based on the column: `calories`.
    - Use the palette: `coolwarm`.

```python
sns.scatterplot(x='potass', y='fiber', data=cereal, hue=cereal['mfr'], size=cereal['calories'], palette='coolwarm')
plt.show()
```

4. Create a scatter plot in Seaborn to visualize the strongest negative correlation you found.

```python
sns.scatterplot(x='rating', y='sugars', data=cereal, hue=cereal['mfr'], size=cereal['calories'], palette='coolwarm')
plt.show()
```
