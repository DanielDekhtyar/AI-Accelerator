**Titanic Dataset:**  

1. Create a new column Age_Category using the **apply** method  
If age is less than 18, assign Child otherwise assign Adult.
2. drop the column you created
3. Create a new column Age_Category using the **apply** method:  
If age is less than 18, assign Child.  
If age is between 18 and 50, assign Adult.  
If age is 60 or more, assign Senior.

```python
titanic = pd.read_csv("titanic.csv")
```

```python
Age_Category = titanic['Age'].apply(
    lambda x: 'Child' if x < 18 else 'Adult'
    ).rename('Age_Category')
```

```python
titanic = pd.concat([titanic, Age_Category], axis=1)
```

```python
titanic
```

```python
titanic = titanic.drop('Age_Category', axis=1)
```

```python
titanic['Age_Category'] = titanic.Age.apply(
    lambda age: "Child" if age < 18 else "Adult" if age < 50 else "Senior")
```

**Cereal Dataset:**  

5. Use the **apply** method to create a new column Health_Score:
Add the values of fiber and protein for each row.

```python
cereal = pd.read_csv("cereal.csv")
```

```python
Health_Score = cereal.apply(lambda x: x.fiber + x.protein, axis=1).rename('Health_Score')
```

```python
Health_Score
```

```python
cereal = pd.concat([cereal, Health_Score], axis=1)
```
