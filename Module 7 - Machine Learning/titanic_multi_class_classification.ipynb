```python
import tensorflow as tf
import seaborn as sns
import pandas as pd
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
#  טעינה
df = sns.load_dataset("titanic")
```

```python
df
```

```python
#  עיבוד נתונים - שמירת משתנים רלוונטיים['pclass', 'sex', 'age', 'fare', 'embark_town']
df = df[['pclass', 'sex', 'age', 'fare', 'embark_town']].dropna()
```

```python
# קידוד משתנים קטגוריים
df = pd.get_dummies(df)
df
```

```python
# הפרדת תכונות ומטרה
X = df.drop(columns=['pclass'])
y = (df['pclass']-1).values
```

```python
# נרמול הנתונים מרשה הפעם לפני ה train_test_split
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
#  פיצול הנתונים לסט אימון וסט בדיקה
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
#ומכאן תחשבו לבד.....
```

```python
y_train = tf.one_hot(y_train, depth=3).numpy()
y_test = tf.one_hot(y_test, depth=3).numpy()
```

```python
model = keras.Sequential([
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(3, activation='softmax')
])
```

```python
model.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.005),
    loss = keras.losses.CategoricalCrossentropy(),
    metrics = ['accuracy']
)
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=16, epochs=100)
```
